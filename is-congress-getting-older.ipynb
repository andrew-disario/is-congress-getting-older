```python
# import libraries
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# path in data
data = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/congress-age/congress-terms.csv')

data
```

```python
# create a dataframe for the house and one for the senate
house = data[data.chamber == 'house']
senate = data[data.chamber == 'senate']
```

```python
# create list for house and populate with each congress and each congress' average age
house_ages = []
for congress in house.congress:
    house_ages.append([congress, house[house.congress == congress].age.mean()])
# create dataframe for house with congress and age as columns
house_ages_df = pd.DataFrame(house_ages, columns = ('congress', 'age'))

# create list for senate and populate with each congress and each congress' average age
senate_ages = []
for congress in senate.congress:
    senate_ages.append([congress, senate[senate.congress == congress].age.mean()])
# create dataframe for senate with congress and age as columns
senate_ages_df = pd.DataFrame(senate_ages, columns = ('congress', 'age'))

senate_ages_df
```

```python
# plot data for house and senate
plt.plot(house_ages_df.congress, house_ages_df.age, color = 'red')
plt.plot(senate_ages_df.congress, senate_ages_df.age, color = 'blue')
plt.legend(['House of Representatives', 'Senate'])
plt.xlabel('Congress')
plt.ylabel('Age')
plt.show()
```

```python
# plot data for house and senate with points
plt.plot(house_ages_df.congress, house_ages_df.age / house_ages_df.age.iloc[0] * 100, 'o-', color = 'red')
plt.plot(senate_ages_df.congress, senate_ages_df.age / senate_ages_df.age.iloc[0] * 100, 'o-', color = 'blue')
plt.legend(['House of Representatives', 'Senate'])
plt.xlabel('Congress')
plt.ylabel('Age')
plt.show()
```
